```python
import pandas as pd
import numpy as np

df = pd.read_csv('ecom.csv')

# Filter data for a specific user
user_id = 1001 
filtered_data = df[df['User_ID'] == user_id]  # Ensure column name matches exactly

# Group purchases by category 
grouped_data = df.groupby('Product_Category')['Purchase_Frequency'].sum().reset_index()

# Identify top-selling products 
top_selling_products = grouped_data.sort_values(by='Purchase_Frequency', ascending=False)

print("Filtered Data for User ID 1001:")
print(filtered_data)
print("\nTop Selling Products:")
print(top_selling_products)
```

```python
import pandas as pd

df = pd.read_csv('clich.csv')

# Function to compute yearly averages
def compute_yearly_averages(dataframe):
    """
    Computes the yearly averages for temperature, humidity, and CO2 levels.
    """
    yearly_avg = dataframe.groupby('Year').mean(numeric_only=True).reset_index()
    return yearly_avg

# Function to identify trends in the data
def identify_trends(dataframe):
    """
    Identifies trends in temperature, humidity, and CO2 levels over the years.
    """
    dataframe = dataframe.set_index("Year")  # Set "Year" as the index before using diff()
    trends = dataframe.diff().mean().to_dict()  # Compute average yearly change for each column
    return trends

# Main function to execute the analysis
def main():
    yearly_averages = compute_yearly_averages(df)
    trends = identify_trends(yearly_averages)
    
    print("Yearly Averages:")
    print(yearly_averages)
    print("\nTrends:")
    print(trends)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the stock price data from the CSV file
df = pd.read_csv('stma.csv')

# Convert "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Function to plot stock performance
def plot_stock_performance(dataframe):
    """
    Plots the stock performance of multiple companies over time.
    """
    plt.figure(figsize=(14, 7))
    
    for company in dataframe['Company'].unique():
        company_data = dataframe[dataframe['Company'] == company].sort_values(by='Date')  # Sort by Date
        plt.plot(company_data['Date'], company_data['Stock_Price'],marker='o', label=company)
    
    plt.title('Stock Performance Comparison')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.xticks(rotation=45)
    plt.grid()
    plt.tight_layout()
    plt.show()

# Main function to execute the visualization
def main():
    plot_stock_performance(df)

if __name__ == "__main__":
    main()
```
